```python
import sys
sys.path.insert(0, '.')

from torch.utils.data import Subset
from main import load_components
```

```python
components = load_components()
trainer = components['trainer']

# subset the train dataset to a handful of samples
N_SAMPLES = 32
small_dataset = Subset(trainer.train_loader.dataset, range(N_SAMPLES))
print(f'Subset size: {len(small_dataset)} samples')
```

```python
from torch.utils.data import DataLoader

trainer.train_loader = DataLoader(small_dataset, batch_size=4, shuffle=True, num_workers=0)
trainer.val_loader = DataLoader(small_dataset, batch_size=4, shuffle=False, num_workers=0)

trainer.train(num_epochs=2)
```
