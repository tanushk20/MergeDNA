```python
import sys
sys.path.insert(0, '..')

from torch.utils.data import DataLoader
from dataloader.dataloader import DNADataset

FASTA_PATH = '../data/human_chr1_4096_filtered.fa'
BATCH_SIZE = 8
```

```python
dataset = DNADataset(FASTA_PATH)
print(f'Sequences: {len(dataset):,}')
print(f'Sample shape: {dataset[0].shape}')  # (4096, 4)
```

```python
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=4)

batch = next(iter(loader))
print(f'Batch shape: {batch.shape}')  # (8, 4096, 4)
print(f'Dtype: {batch.dtype}')
print(f'Min/max: {batch.min().item():.0f} / {batch.max().item():.0f}')
```
